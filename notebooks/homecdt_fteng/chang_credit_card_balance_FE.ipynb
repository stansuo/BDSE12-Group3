```python
def total_info(data):
    
    #＊要改columns檔名
    data_shape=pd.DataFrame([data.shape[0],data.shape[1],data.size],index=['Row','Column','Total Data'],columns=['credit_card_balance'])
    data_type_counts=pd.DataFrame(data.dtypes.value_counts(),columns=['credit_card_balance'])
    data_shape_df=data_shape.append(data_type_counts)
    print(data_shape_df)
  
    #a欄位型態
    data_type=data.dtypes
    
    #b差異值數量
    uniques_num=data.apply(lambda x:x.unique().shape[0],axis=0)
    
    #c空值比例
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    
    #abc
    columns_data=pd.concat([data_type,uniques_num,total, percent], axis=1, keys=['Type','Uniques_Num','Missing_Total', 'Missing_Percent'],sort=False)
    return columns_data
    #return data_shape_df,columns_data
    
```

```python
total_info(credit_card_balance)
```

### <a id=012>資料的空值分佈</a>

```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    print('missing data')
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
def columns_object(data):
    columns_object= data.columns[data.dtypes == 'object']
    columns_object_pd =pd.DataFrame(columns_object,columns=['類別型欄位'])

    unique_object = data.select_dtypes(include=['object']).nunique()
    unique_object_pd=pd.DataFrame(unique_object,columns=['Unique_Num'])
    
    print('類別型欄位計共 {} 個。'.format(len(data.select_dtypes(include=['object']).columns)))
    print('')
    
    print(pd.merge(columns_object_pd,unique_object_pd,left_on='類別型欄位',right_index=True))
```

### <a id=031>原資料的Heatmap</a>

```python
def heatmap(data):
    data_corr=data.corr()
    plt.figure(figsize=(15,15),dpi=80)
    sns.heatmap(np.abs(data.corr()),cmap="Blues",
               vmin=0,vmax=1,square=True,annot=True,fmt='.2f')
    plt.show()
    print('數值型別欄位計共 {} 個。'.format(len(data.select_dtypes(include=['int64','float64']).columns)))
    return(data_corr)
```

```python
ccb = credit_card_balance.sort_values(by=['SK_ID_CURR','SK_ID_PREV','MONTHS_BALANCE']).copy()

# 新增欄位
#1、需要還款有幾個月份(COUNT AMT_TOTAL_RECEIVABLE > 0) 1為有還款，0為沒還款
ccb['COUNT_AMT_TOTAL_RECEIVABLE']=(ccb['AMT_TOTAL_RECEIVABLE']>0).replace(True,1)

##2、各管道借錢佔額度的比例平均、最大、最小:
ccb['AMT_DRAWINGS_ATM_SUM']=ccb['AMT_DRAWINGS_ATM_CURRENT'] #ATM借錢金額合計
ccb['AMT_DRAWINGS_ATM_MAX']=ccb['AMT_DRAWINGS_ATM_CURRENT'] #ATM借錢最大金額
#ATM借錢最小金額，有踢除0
ccb['AMT_DRAWINGS_ATM_MIN']=ccb['AMT_DRAWINGS_ATM_CURRENT'].copy()
ccb['AMT_DRAWINGS_ATM_MIN'].replace(0,np.nan,inplace=True)

ccb['AMT_DRAWINGS_OTHER_SUM']=ccb['AMT_DRAWINGS_OTHER_CURRENT'] #OTHER借錢金額合計
ccb['AMT_DRAWINGS_OTHER_MAX']=ccb['AMT_DRAWINGS_OTHER_CURRENT'] #OTHER借錢最大金額
#OTHER借錢最小金額，有踢除0
ccb['AMT_DRAWINGS_OTHER_MIN']=ccb['AMT_DRAWINGS_OTHER_CURRENT'].copy()
ccb['AMT_DRAWINGS_OTHER_MIN'].replace(0,np.nan,inplace=True)

ccb['AMT_DRAWINGS_POS_SUM']=ccb['AMT_DRAWINGS_OTHER_CURRENT'] #POS借錢金額合計
ccb['AMT_DRAWINGS_POS_MAX']=ccb['AMT_DRAWINGS_OTHER_CURRENT'] #POS借錢最大金額
#POS借錢最小金額，有踢除0
ccb['AMT_DRAWINGS_POS_MIN']=ccb['AMT_DRAWINGS_OTHER_CURRENT'].copy()
ccb['AMT_DRAWINGS_POS_MIN'].replace(0,np.nan,inplace=True)


#4、可用額度金額(CREDIT_LIMIT MAX,MIN,MEDIAN,MOST RECENT,FAREST)(AMT_CREDIT_LIMIT_ACTUAL):
ccb['CREDIT_LIMIT_MAX']=ccb['AMT_CREDIT_LIMIT_ACTUAL'] #可用最大額度金額
ccb['CREDIT_LIMIT_MIN']=ccb['AMT_CREDIT_LIMIT_ACTUAL'] #可用最小額度金額
ccb['CREDIT_LIMIT_MEDIAN']=ccb['AMT_CREDIT_LIMIT_ACTUAL'] #可用中位數額度金額 
ccb['CREDIT_LIMIT_MOST_RECENT'] = ccb['AMT_CREDIT_LIMIT_ACTUAL'] #最近可用額度金額
ccb['CREDIT_LIMIT_FAREST'] = ccb['AMT_CREDIT_LIMIT_ACTUAL']  #最久之前可用額度金額






ccb['MONTHS_BALANCE_start']=ccb['MONTHS_BALANCE']    # 8、紀錄時間最久以前是多久
ccb['MONTHS_BALANCE_near']=ccb['MONTHS_BALANCE']    # 9、紀錄時間最近是幾個月以前
ccb['AMT_TOTAL_RECEIVABLE_MAX']=ccb['AMT_TOTAL_RECEIVABLE']      # 13、最大總借款金額
ccb['AMT_DRAWINGS_CURRENT_MAX']=ccb['AMT_DRAWINGS_CURRENT'] # 14、最大借款金額

#16、最小借款金額，有踢除0
ccb['AMT_DRAWINGS_CURRENT_MIN']=ccb['AMT_DRAWINGS_CURRENT'].copy()
ccb['AMT_DRAWINGS_CURRENT_MIN'].replace(0,np.nan,inplace=True)

ccb['AMT_RECEIVABLE_PRINCIPAL_MAX']=ccb['AMT_RECEIVABLE_PRINCIPAL']#最大RECEIVABLE PRINCIPAL動用額度佔比
ccb['AMT_TOTAL_RECEIVABLE_MAX']=ccb['AMT_TOTAL_RECEIVABLE_MAX']#最大AMT_TOTAL_RECEIVABLE動用額度佔比

ccb['AMT_DRAWINGS_CURRENT_SUM']=ccb['AMT_DRAWINGS_CURRENT'] # 合計借款金額
ccb['CNT_DRAWINGS_CURRENT_SUM']=ccb['CNT_DRAWINGS_CURRENT'] # 合計借款次數



# 創造新的欄位 
num_aggregations = {
        'COUNT_AMT_TOTAL_RECEIVABLE':'sum',
        'AMT_DRAWINGS_ATM_SUM':'sum',
        'AMT_DRAWINGS_ATM_MAX':'max',
        'AMT_DRAWINGS_ATM_MIN':'min',
        'AMT_DRAWINGS_OTHER_SUM':'sum',
        'AMT_DRAWINGS_OTHER_MAX':'max',
        'AMT_DRAWINGS_OTHER_MIN':'min',
        'AMT_DRAWINGS_POS_SUM':'sum',
        'AMT_DRAWINGS_POS_MAX':'max',
        'AMT_DRAWINGS_POS_MIN':'min',
        'CREDIT_LIMIT_MAX':'max',
        'CREDIT_LIMIT_MIN':'min',
        'CREDIT_LIMIT_MEDIAN':'median',
        'CREDIT_LIMIT_MOST_RECENT':'last',
        'CREDIT_LIMIT_FAREST':'first',
        'MONTHS_BALANCE_start':'min',
        'MONTHS_BALANCE_near':'max',
        'AMT_TOTAL_RECEIVABLE_MAX':'max',
        'AMT_DRAWINGS_CURRENT_MAX':'max',
        'AMT_DRAWINGS_CURRENT_MIN':'min',
        'AMT_RECEIVABLE_PRINCIPAL_MAX':'max',
        'AMT_TOTAL_RECEIVABLE_MAX':'max',
        'AMT_DRAWINGS_CURRENT_SUM':'sum',
        'CNT_DRAWINGS_CURRENT_SUM':'sum'
    
        
        
        
    }

ccb_1 = ccb.groupby(['SK_ID_CURR','SK_ID_PREV']).agg({**num_aggregations})
                                    
# # 重新排列
# ccb_1.reset_index(level=('SK_ID_CURR',
#                           'SK_ID_PREV'),inplace=True)







#2、各管道借錢佔額度的比例平均、最大、最小:
#ATM借錢佔額度的比例平均
ccb_1['AMT_DRAWINGS_ATM_SUM_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_ATM_SUM']/ccb_1['CREDIT_LIMIT_MEDIAN']
#ATM借錢最大金額佔額度的比例
ccb_1['AMT_DRAWINGS_ATM_MAX_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_ATM_MAX']/ccb_1['CREDIT_LIMIT_MEDIAN']
#ATM借錢最小金額佔額度的比例
ccb_1['AMT_DRAWINGS_ATM_MIN_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_ATM_MIN']/ccb_1['CREDIT_LIMIT_MEDIAN']

#OTHER借錢佔額度的比例平均
ccb_1['AMT_DRAWINGS_OTHER_SUM_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_OTHER_SUM']/ccb_1['CREDIT_LIMIT_MEDIAN']
#OTHER借錢最大金額佔額度的比例
ccb_1['AMT_DRAWINGS_OTHER_MAX_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_OTHER_MAX']/ccb_1['CREDIT_LIMIT_MEDIAN']
#OTHER借錢最小金額佔額度的比例
ccb_1['AMT_DRAWINGS_OTHER_MIN_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_OTHER_MIN']/ccb_1['CREDIT_LIMIT_MEDIAN']

#POS借錢佔額度的比例平均
ccb_1['AMT_DRAWINGS_POS_SUM_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_POS_SUM']/ccb_1['CREDIT_LIMIT_MEDIAN']
#POS借錢最大金額佔額度的比例
ccb_1['AMT_DRAWINGS_POS_MAX_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_POS_MAX']/ccb_1['CREDIT_LIMIT_MEDIAN']
#POS借錢最小金額佔額度的比例
ccb_1['AMT_DRAWINGS_POS_MIN_CREDIT_LIMIT_MEDIAN _rate']=ccb_1['AMT_DRAWINGS_POS_MIN']/ccb_1['CREDIT_LIMIT_MEDIAN']

#3、各管道借錢佔總借款額的比例(每一期？) (xx/AMT_DRAWINGS_CURRENT)
# ATM借錢佔總借款額的比例平均
ccb_1['AMT_DRAWINGS_ATM_ALL_DRAWINGS_rate']=ccb['AMT_DRAWINGS_ATM_CURRENT']/ccb['AMT_DRAWINGS_CURRENT']
#OTHER借錢佔總借款額的比例平均
ccb_1['AMT_DRAWINGS_OTHER_ALL_DRAWINGS_rate']=ccb['AMT_DRAWINGS_OTHER_CURRENT']/ccb['AMT_DRAWINGS_CURRENT']
#POS借錢佔總借款額的比例平均
ccb_1['AMT_DRAWINGS_POS_ALL_DRAWINGS_rate']=ccb['AMT_DRAWINGS_POS_CURRENT']/ccb['AMT_DRAWINGS_CURRENT']


##5、最大動用額度佔比(RECEIVABLE PRINCIPAL/CREDIT LIMIT)、(AMT_TOTAL_RECEIVABLE/CREDIT LIMIT):
#最大RECEIVABLE PRINCIPAL動用額度佔比(RECEIVABLE PRINCIPAL/CREDIT LIMIT)
ccb_1['RECEIVABLE PRINCIPAL_rate']=ccb_1['AMT_RECEIVABLE_PRINCIPAL_MAX']/ccb_1['CREDIT_LIMIT_MAX'] 
#最大AMT_TOTAL_RECEIVABLE動用額度佔比(AMT_TOTAL_RECEIVABLE/CREDIT LIMIT)
ccb_1['AMT_TOTAL_RECEIVABLE_rate']=ccb_1['AMT_TOTAL_RECEIVABLE_MAX']/ccb_1['CREDIT_LIMIT_MAX']


#12、每筆貸款規定償還金額占借錢% (MIN_REGULARITY/TOTAL_RECEIVABLE)->(AMT_INST_MIN_REGULARITY/AMT_TOTAL_RECEIVABLE)
#ccb_1['MIN_REGULARITY_TOTAL_RECEIVABLE_rate']=ccb['AMT_INST_MIN_REGULARITY']/ccb['AMT_TOTAL_RECEIVABLE']???


#15、最大借款金額占額度%(AMT_DRAWINGS_CURRENT/CREDIT_LIMIT MAX)
#AMT_DRAWINGS_CURRENT_MAX_rate --最大借款金額占額度%
ccb_1['CREDIT_LIMIT_MAX_rate']=ccb_1['AMT_DRAWINGS_CURRENT_MAX']/ccb_1['CREDIT_LIMIT_MAX']

#17、最小借款金額占額度%(AMT_DRAWINGS_CURRENT/CREDIT_LIMIT MIN)
ccb_1['CREDIT_LIMIT_MIN_rate_MAX']=ccb_1['AMT_DRAWINGS_CURRENT_MIN']/ccb_1['CREDIT_LIMIT_MAX']
ccb_1['CREDIT_LIMIT_MIN_rate_MIN']=ccb_1['AMT_DRAWINGS_CURRENT_MIN']/ccb_1['CREDIT_LIMIT_MIN']
ccb_1['CREDIT_LIMIT_MIN_rate_MEDIAN']=ccb_1['AMT_DRAWINGS_CURRENT_MIN']/ccb_1['CREDIT_LIMIT_MEDIAN']

#18、貸款借錢時平均每次提領多少(AMT_DRAWINGS_CURRENT/CNT_DRAWINGS_CURRENT MEAN)
ccb_1['AMT_DRAWINGS_CURRENT_CNT_rate']=ccb_1['AMT_DRAWINGS_CURRENT_SUM']/ccb_1['CNT_DRAWINGS_CURRENT_SUM']

#19、每筆貸款借錢時平均每次提領多少占額度的% ((AMT_DRAWINGS_CURRENT/CNT_DRAWINGS_CURRENT MEAN)/CREDIT_LIMIT)
ccb_1['AMT_DRAWINGS_CURRENT_CNT_CREDIT_LIMIT_rate']=ccb_1['AMT_DRAWINGS_CURRENT_SUM']/ccb_1['CNT_DRAWINGS_CURRENT_SUM']/ccb_1['CREDIT_LIMIT_MEDIAN']
        

ccb_1.shape  # (104307, 43)   
ccb_1.head(10)
```

```python
ccb4 = credit_card_balance[credit_card_balance['AMT_INST_MIN_REGULARITY']>0].sort_values(by=['SK_ID_CURR','SK_ID_PREV','MONTHS_BALANCE']).copy()

# 7、有借錢時，平均一期還應還金額的多少%(AMT_INST_MIN_REGULARITY/AMT_PAYMENT_TOTAL_CURRENT)
# (AMT_PAYMENT_TOTAL_CURRENT/AMT_INST_MIN_REGULARITY)(要剔除AMT_INST_MIN_REGULARITY是0的)
#AMT_PAYMENT_TOTAL_CURRENT_MIN_REGULARITY_rate -->有借錢時，AMT_PAYMENT_TOTAL_CURRENT平均一期還應還MIN_REGULARITY金額的多少%

# (AMT_PAYMENT_CURRENT/AMT_INST_MIN_REGULARITY)(要剔除AMT_INST_MIN_REGULARITY是0的)
#AMT_PAYMENT_CURRENT_MIN_REGULARITY_rate -->有借錢時，AMT_PAYMENT_CURRENT平均一期還應還MIN_REGULARITY金額的多少%


# 新增欄位

ccb4['AMT_INST_MIN_REGULARITY_count']=ccb4['AMT_INST_MIN_REGULARITY']
ccb4['AMT_PAYMENT_TOTAL_MIN_REGULARITY_rate']=ccb4['AMT_PAYMENT_TOTAL_CURRENT']/ccb4['AMT_INST_MIN_REGULARITY']
ccb4['AMT_PAYMENT_MIN_REGULARITY_rate']=ccb4['AMT_PAYMENT_CURRENT']/ccb4['AMT_INST_MIN_REGULARITY']

# 創造新的欄位 
num_aggregations = {
        'AMT_INST_MIN_REGULARITY_count':'count',
        'AMT_PAYMENT_TOTAL_MIN_REGULARITY_rate':'sum',
        'AMT_PAYMENT_MIN_REGULARITY_rate':'sum'
        
    }


ccb4_1 = ccb4.groupby(['SK_ID_CURR','SK_ID_PREV']).agg({**num_aggregations})
                                    
# # 重新排列
# ccb4_1.reset_index(level=('SK_ID_CURR',
#                           'SK_ID_PREV'),inplace=True)



ccb4_1['AMT_PAYMENT_TOTAL_MIN_REGULARITY_mean_rate']=ccb4_1['AMT_PAYMENT_TOTAL_MIN_REGULARITY_rate']/ccb4_1['AMT_INST_MIN_REGULARITY_count']
ccb4_1['AMT_PAYMENT_MIN_REGULARITY_mean_rate']=ccb4_1['AMT_PAYMENT_MIN_REGULARITY_rate']/ccb4_1['AMT_INST_MIN_REGULARITY_count']


# ccb4_1.shape  #(70179, 5)
ccb4_1.head(33)
```

```python
# 10、有借錢紀錄時間最久是多久以前
# AMT_BALANCE_MONTHS_start --> AMT_BALANCE紀錄開始時間

# 11、有借錢紀錄時間最近是多久以前
# AMT_BALANCE_MONTHS_near --> AMT_BALANCE紀錄最近時間

ccb2 = credit_card_balance[credit_card_balance['CNT_DRAWINGS_CURRENT']!=0].sort_values(by=['SK_ID_CURR','SK_ID_PREV','MONTHS_BALANCE']).copy()
ccb2['AMT_BALANCE_MONTHS_start'] = ccb2['MONTHS_BALANCE']
ccb2['AMT_BALANCE_MONTHS_near'] = ccb2['MONTHS_BALANCE']

ccb2_1 = ccb2.groupby(['SK_ID_CURR','SK_ID_PREV']).agg({'AMT_BALANCE_MONTHS_start':'min','AMT_BALANCE_MONTHS_near':'max',})
                                    
# # 重新排列
# ccb2_1.reset_index(level=('SK_ID_CURR',
#                           'SK_ID_PREV'),inplace=True)


# ccb2_1.shape  #(69988, 2)
ccb2_1.head()
```

* a1:新增7個欄位：將同一個'SK_ID_PREV'彙整，列出NAME_CONTRACT_STATUS(先前信用期間合約狀態)發生的比例。併到credit_card_balance。
* a2:新增2個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD(先前信用期間逾期天數)發生的各項統計('max'(最多天),'mean'(平均天數))。
* a3:新增1個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD(先前信用期間逾期天數)發生的次數。
* a4:新增2個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD_DEF(先前信用期間逾期天數(可容忍,忽略低金額))發生的各項統計('max'(最多天),'mean'(平均天數))。]
* a5:新增1個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD_DEF(先前信用期間逾期天數(可容忍,忽略低金額))發生的次數。

```python
ccb3 = credit_card_balance.sort_values(by=['SK_ID_CURR','SK_ID_PREV','MONTHS_BALANCE']).copy()

#a1:新增7個欄位：將同一個'SK_ID_PREV'彙整，列出NAME_CONTRACT_STATUS(先前信用期間合約狀態)發生的比例。
# 將有特殊狀態的欄位新增 (onehot)
categorical_columns = []
for name in ccb3:
    if ccb3[name].dtype=='object':
        categorical_columns.append(name)
ccb3 = pd.get_dummies(ccb3, columns = categorical_columns)

ccb3.rename(columns={'NAME_CONTRACT_STATUS_Active':'CREDIT_ACTIVE_Active_rate',
                     'NAME_CONTRACT_STATUS_Approved':'CREDIT_ACTIVE_Approved_rate',
                     'NAME_CONTRACT_STATUS_Completed':'CREDIT_ACTIVE_Completed_rate',
                     'NAME_CONTRACT_STATUS_Demand':'CREDIT_ACTIVE_Demand_rate',
                     'NAME_CONTRACT_STATUS_Refused':'CREDIT_ACTIVE_Refused_rate',
                     'NAME_CONTRACT_STATUS_Sent proposal':'CREDIT_ACTIVE_Sent proposal_rate',
                     'NAME_CONTRACT_STATUS_Signed':'CREDIT_ACTIVE_Signed_rate'}
                     ,inplace=True)


# 新增欄位

#a2:新增2個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD(先前信用期間逾期天數)發生的各項統計('max'(最多天),'mean'(平均天數))。
ccb3['ccb_SK_DPD_MAX']=ccb3['SK_DPD'] # SK_DPD逾期天數最多天
ccb3['ccb_SK_DPD_MEAN']=ccb3['SK_DPD'] # SK_DPD逾期天數平均天數

#a3:新增1個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD(先前信用期間逾期天數)發生的次數。
#新增'SK_DPD_delay'欄位判斷'SK_DPD'是否有違約過，有發生過的顯示1，沒有的顯示0
#將逾期次數相加
ccb3['ccb_SK_DPD_delay_count']=(ccb3['SK_DPD']>0).replace(True,1)

#a4:新增2個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD_DEF(先前信用期間逾期天數(可容忍,忽略低金額))發生的各項統計('max'(最多天),'mean'(平均天數))。
ccb3['ccb_SK_DPD_DEF_MAX']=ccb3['SK_DPD_DEF'] # SK_DPD_DEF逾期天數最多天
ccb3['ccb_SK_DPD_DEF_MEAN']=ccb3['SK_DPD_DEF'] # SK_DPD_DEF逾期天數平均天數

#a5:新增1個欄位：將同一個'SK_ID_PREV'彙整，列出SK_DPD_DEF(先前信用期間逾期天數(可容忍,忽略低金額))發生的次數。
#新增'SK_DPD_DEF_delay'欄位判斷'SK_DPD'是否有違約過，有發生過的顯示1，沒有的顯示0
#將逾期次數相加
ccb3['ccb_SK_DPD_DEF_delay_count']=(ccb3['SK_DPD_DEF']>0).replace(True,1)



# 創造新的欄位 
num_aggregations = {
        'MONTHS_BALANCE':'count',
        'CREDIT_ACTIVE_Active_rate':'sum',
        'CREDIT_ACTIVE_Approved_rate':'sum',
        'CREDIT_ACTIVE_Completed_rate':'sum',
        'CREDIT_ACTIVE_Demand_rate':'sum',
        'CREDIT_ACTIVE_Refused_rate':'sum',
        'CREDIT_ACTIVE_Sent proposal_rate':'sum',
        'CREDIT_ACTIVE_Signed_rate':'sum',
        'ccb_SK_DPD_MAX':'max',
        'ccb_SK_DPD_MEAN':'mean',
        'ccb_SK_DPD_delay_count':'sum',
        'ccb_SK_DPD_DEF_MAX':'max',
        'ccb_SK_DPD_DEF_MEAN':'mean',
        'ccb_SK_DPD_DEF_delay_count':'sum'
    
    
    }

ccb3_1 = ccb3.groupby(['SK_ID_CURR','SK_ID_PREV']).agg({**num_aggregations})

# 做更名
ccb3_1.rename(columns={'MONTHS_BALANCE':'Total_Months'
                     },inplace=True)

# # 重新排列
# ccb3_1.reset_index(level=('SK_ID_CURR',
#                           'SK_ID_PREV'),inplace=True)

#a1:新增7個欄位：將同一個'SK_ID_PREV'彙整，列出NAME_CONTRACT_STATUS(先前信用期間合約狀態)發生的比例。
# # 狀態發生期數除以總還款期數

ccb3_1['CREDIT_ACTIVE_Active_rate'] = ccb3_1['CREDIT_ACTIVE_Active_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Approved_rate'] = ccb3_1['CREDIT_ACTIVE_Approved_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Completed_rate'] = ccb3_1['CREDIT_ACTIVE_Completed_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Demand_rate'] = ccb3_1['CREDIT_ACTIVE_Demand_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Refused_rate'] = ccb3_1['CREDIT_ACTIVE_Refused_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Sent proposal_rate'] = ccb3_1['CREDIT_ACTIVE_Sent proposal_rate']/ccb3_1['Total_Months']
ccb3_1['CREDIT_ACTIVE_Signed_rate'] = ccb3_1['CREDIT_ACTIVE_Signed_rate']/ccb3_1['Total_Months']



ccb3_1.shape  #(104307, 14)
ccb3_1.head(20)
```
